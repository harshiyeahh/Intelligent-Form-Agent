# Intelligent Form Agent - Exploratory Notebook

This notebook demonstrates how to use the core methods of the Intelligent Form Agent to load, parse, summarize, and query documents.

```python
import os
import sys
from dotenv import load_dotenv

# Ensure project root is in path
sys.path.insert(0, os.path.abspath(".."))
load_dotenv()

from src.core import extract_structured_data, ask_question, summarize_form, analyze_multiple_forms
```

```python
# Check if API Key is configured
api_key = os.getenv("GEMINI_API_KEY")
if not api_key:
    print("WARNING: GEMINI_API_KEY not found in environment. Please add it to your .env file or configure it.")
else:
    print("Gemini API key configured successfully.")
```

```python
# 1. Run Structured JSON Extraction
invoice_path = "../data/sample_invoice.png"
if os.path.exists(invoice_path):
    try:
        data = extract_structured_data(invoice_path)
        import json
        print(json.dumps(data, indent=2))
    except Exception as e:
        print(f"Extraction error: {e}")
```

```python
# 2. Single Document Q&A
medical_path = "../data/sample_medical.png"
if os.path.exists(medical_path):
    try:
        answer = ask_question(medical_path, "What medications is the patient currently taking?")
        print(f"Q: What medications is the patient currently taking?\nA: {answer}")
    except Exception as e:
        print(f"Q&A error: {e}")
```

```python
# 3. Summarization
rental_path = "../data/sample_rental.png"
if os.path.exists(rental_path):
    try:
        summary = summarize_form(rental_path)
        print(summary)
    except Exception as e:
        print(f"Summarization error: {e}")
```

```python
# 4. Multi-document comparison
files = ["../data/sample_invoice.png", "../data/sample_rental.png"]
if all(os.path.exists(f) for f in files):
    try:
        query = "Create a markdown table comparing the main active parties and total monetary amounts in these documents."
        comparison = analyze_multiple_forms(files, query)
        print(comparison)
    except Exception as e:
        print(f"Comparison error: {e}")
```
